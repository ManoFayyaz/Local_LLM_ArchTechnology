Medical Dataset

```python
from datasets import load_dataset

# 1. Load a medical dataset (using 1000 rows for speed, as per Kaggle best practices)
dataset = load_dataset("ruslanmv/ai-medical-chatbot", split="train[:1000]")

# 2. Define the Medical Prompt Template
medical_prompt = """### Instruction:
You are a knowledgeable medical assistant. Provide a concise and accurate response to the patient's query.

### Patient Query:
{}

### Doctor's Response:
{}"""

EOS_TOKEN = tokenizer.eos_token

def format_prompts(examples):
    instructions = examples["Patient"]
    outputs = examples["Doctor"]
    texts = []
    for instruction, output in zip(instructions, outputs):
        # Join instruction and output into the template + add EOS token
        text = medical_prompt.format(instruction, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts }

# 3. Map the formatting to the dataset
dataset = dataset.map(format_prompts, batched=True)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

# Apply LoRA (Low-Rank Adaptation) to make training fast
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
)

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = 2048,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        report_to = "none",  # Prevents logging hangs
        warmup_steps = 5,
        max_steps = 60, # Exactly 60 steps as requested!
        learning_rate = 2e-4,
        fp16 = True,
        bf16 = False,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
    ),
)

print("Trainer is ready! Run trainer.train() next.")
```

```python
trainer.train() 
```

Testing the fintuned model

```python
# 1. Switch model to 'Inference' mode (makes it faster for answering)
FastLanguageModel.for_inference(model) 

# 2. Define a medical question
inputs = tokenizer(
[
    medical_prompt.format(
        "A 45-year-old patient presents with sudden chest pain and shortness of breath. What are the top 3 differential diagnoses?", # Question
        "", # Leave Reasoning empty for the model to fill
        "", # Leave Response empty for the model to fill
    )
], return_tensors = "pt").to("cuda")

# 3. Generate the answer
outputs = model.generate(**inputs, max_new_tokens = 500, use_cache = True)
response = tokenizer.batch_decode(outputs)

# 4. Print the result
print(response[0])
```

```python
import os
import zipfile
from IPython.display import FileLink

# 1. Zip the folder
def zip_dir(directory, file_name):
    with zipfile.ZipFile(file_name, 'w', zipfile.ZIP_DEFLATED) as zip_ref:
        for root, dirs, files in os.walk(directory):
            for file in files:
                zip_ref.write(os.path.join(root, file), 
                              os.path.relpath(os.path.join(root, file), 
                              os.path.join(directory, '..')))

zip_dir('medical_model', 'my_medical_model.zip')

# 2. Create the download link
FileLink(r'my_medical_model.zip')
```
